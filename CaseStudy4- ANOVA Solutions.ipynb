XYZ Company has offices in four different zones. The company wishes to
investigate the following :
.

● The mean sales generated by each zone.


● Total sales generated by all the zones for each month.


● Check whether all the zones generate the same amount of sales.


Help the company to carry out their study with the help of data provided

```python
import numpy as no
import pandas as pd
import matplotlib.pyplot as mlt
import seaborn as sns
from scipy import stats
from scipy.stats import f_oneway
```

```python
df=pd.read_csv("Sales_Data_zone_wise.csv")
df
```

```python
print("Mean of each zones")
df[["Zone - A","Zone - B","Zone - C","Zone - D"]].mean()
```

```python
#we can plot a box plot to check the total sales by all zones in each month. 
```

```python
bp=df.boxplot(showmeans=True)
bp.set_title("BoxPlot of Zones vs Sales", fontsize=20, color='green')
mlt.xlabel("Zones")
mlt.ylabel("Sales")
```

```python
#Total sales generated by all the zones for each month
total=df.mean(axis=1)
total
```

```python
mlt.figure(figsize=(15,8))
mlt.plot(total)
```

```python
# Check whether all the zones generate the same amount of sales.
```

```python
H0: All zones generate same amount of sales 
H0: Atleast one zone generate different amount of sales
```

```python
F_value, p_val = stats.f_oneway(df['Zone - A'], df['Zone - B'],df['Zone - C'], df['Zone - D'])
print("F-values: ",F_value,"\np values: ",p_val)
alpha=0.05
if p_val>alpha:
    print("We fail to reject the Null Hypothesis and hence all the zones generate same amount of sales")
else:
    print("The Null Hypothesis is Rejected and e can concluse that atleast one zone is generating different amount of sales")
```
